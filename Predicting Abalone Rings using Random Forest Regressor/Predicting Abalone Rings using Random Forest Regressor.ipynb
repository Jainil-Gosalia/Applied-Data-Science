```python
#Libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestRegressor
from sklearn.datasets import make_regression
from sklearn import preprocessing
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import explained_variance_score
from sklearn.metrics import r2_score
```

```python
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/abalone/abalone.data"
labels = ["sex","length","diameter","height","whole weight","shucked weight",
          "viscera weight","shell weight","rings"]
attribute_type = {'sex':'category','length':'float','diameter':'float',
                 'height':'float','whole weight':'float','shucker weight':'float',
                 'viscera weight':'float','shell weight':'float','rings':'int'}
df = pd.read_csv(url, names = labels, dtype=attribute_type)
```

```python
#Copying df to avoid manipulating original data
df2 = df.copy()

df2['sex'].replace({'M', 'F','I'},{-1, 0, 1}, inplace=True)
```

```python
df2.head()
```

```python
#Preprocessing
X = df2.iloc[:, :-1].values
y = df2.iloc[:, 7].values

#Split testing and training data in 70 to 30
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
#Random Forest Approach
regr = RandomForestRegressor(max_depth=5, random_state=0)
regr.fit(X_train, y_train)
y_pred = regr.predict(X_test)
```

```python
#Results
explained_variance_score(y_test, y_pred)
r = r2_score(y_test, y_pred)
```

```python
print("The model is : " + str(r*100) + "% accurate from testing")
```
